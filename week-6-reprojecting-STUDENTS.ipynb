# reprojecting or something

to beo onth esafe side, we will reproject to make sure both datasets have the same CRS.

**Reprojectsing** means changing coordinate reference systems of your geospatial data

changing crs in geopandas is super easy using th efollowing:
    
```python
updated_geodf = geodf.to_crs(new_crs)
```

where:
    
- `updated_geodf` is the output of the method, a new geodataframe (to_crs() does not work in place),
- `geodf` is the geopandas.GeoDataFrame we want to transform,
- `new_crs` the CRS we want to convert to, this is an object of type CRS or string representing the CRS (ex: 'epsg:3413') 

````python
countries = countries.to_crs(communities.crs)

# Verify CRSs match
assert countries.crs == communities.crs

fig, ax = plt.subplots()
countries.plot(ax=ax)
communities.plot(ax=ax, color='red')
plt.show()
````

````python
# create a dataframe with the number of communieis per country
n_comms = (communities.groupby('country')
                      .size()
                      .reset_index(name = 'n_communities'))
````

To merge:

```python
output_df = pd.merge(left_df,
                     right_df,
                     how = type_of_join,
                     on = column_to_join)
```

where:
- `output_df` = df resulting from merge
- `left_df` = df on the 'left side'
- `right_df` = df on 'right side'
- `how` = type of join
- `on` = column to join on, must be in both dfs

````python
n_comms = pd.merge(n_comms,
                   country_names,
                   how='left',
                   on='country')
n_comms
````

````python
# merge the data frames
arctic_countries = pd.merge(countries,
                            n_comms,
                            how = 'inner',
                            on = 'admin')
arctic_countries
````

when doing the inner join we dropped the alad islands form the n_comms dataset

````python
# check alan islands is nowhere in countries data frame
'Aland Islands' in countries.values
````

````python
arctic_countries = arctic_countries.set_index('admin')
````

````python
# .at work as ['index', 'column so source value']
# += operator just adds a number to a variable, ex: a=0, a += 1, then a = 1

arctic_countries.at['Finland', 'n_communities'] += 1
````

````python
arctic_countries.plot(column = 'n_communities',
                      legend = True)
````

### reproject to crs optimized for the arctic region

````python
# Reproject to CRS optimized for Arctic region
arctic_countries = arctic_countries.to_crs('epsg:3413')

arctic_countries.plot(column = 'n_communities',
                      legend = True)
````
